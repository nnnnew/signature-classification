# Import modules

```python
from sklearn.model_selection import train_test_split

from skimage.io import imsave

from matplotlib import pyplot as plt
import sys, os, glob
import h5py
import matplotlib.pyplot as plt
%matplotlib inline
```

# Define params and helpers

```python
# Train image and train label define variable
train_image_path = 'train_images.h5'
train_label_path = 'train_labels.h5'
train_image_key = 'NISDCC-offline'
train_label_key = 'NISDCC-offline-labels'

path_data_train = './data/train/'
path_data_test = './data/test/'
```

# Read images and labels

```python
with h5py.File(train_image_path, 'r') as hf:
    x = hf[train_image_key][:]
```

```python
with h5py.File(train_label_path, 'r') as hf:
    y = hf[train_label_key][:]
    y = y-1
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x[i], cmap='gray')
    plt.xlabel(y[i])
```

# Split train and test

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
print('Num of x_train is ' + str(len(x_train)))
print('Num of x_test is ' + str(len(x_test)))
```

# Save image

```python
for i in range(len(x_train)):
    label = y_train[i]
    if not os.path.exists(path_data_train + str(label)):
        os.mkdir(path_data_train + str(label))
    path_image_i = path_data_train + str(label) + '/'
    imsave(path_image_i + str(i) + '.png', x_train[i].astype(int))
```

```python
for i in range(len(x_test)):
    label = y_test[i]
    if not os.path.exists(path_data_test + str(label)):
        os.mkdir(path_data_test + str(label))
    path_image_i = path_data_test + str(label) + '/'
    imsave(path_image_i + str(i) + '.png', x_test[i].astype(int))
```
